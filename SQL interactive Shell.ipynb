## CREATE DATABASE AND INSERT THE VALUE 

```python
import pandas as pd
from pandas import DataFrame
import sqlite3
import re

df = pd.read_csv('roomAllocations.csv', na_values=["-"])


############################# Instructor Data #############################
instructorName = []
instructorID = []
for i in df['Instructor']:
    if i not in instructorName:
        instructorName.append(i)
for i,j in enumerate(instructorName,1):
    instructorID.append(i)
instructorDict = {'INSTRUCTORID':instructorID, 'INSTRUCTORNAME':instructorName}


############################ Course Data ##################################
courseID = []
courseCode= []
courseName = [k for k in df['CourseName']]
for i,j in enumerate(df['CourseName'],1):
    courseID.append(i)
    courseCode.append(j)

courseDict = {'COURSEID':courseID,'COURSECODE':courseCode,'COURSENAME':courseName}

########################### ROOM DATA #######################################

roomName = [r for r in df['RoomName']]
instructorID123 = []
anotherInstructorDict = dict(zip(instructorName, instructorID))
for i in df['Instructor']:  
    instructorID123.append(anotherInstructorDict[i])
days = [d for d in df['Days']]

roomDict = {'COURSEID':courseID, 'INSTRUCTORID':instructorID123, 'ROOMNAME':roomName, 'DAYS':days}

########################### TIME DATA #######################################

df['OfferTimeFrom'] =  pd.to_datetime(df['OfferTimeFrom'], format ='%I:%M %p').dt.strftime('%H:%M')
df['OfferTimeTo'] =  pd.to_datetime(df['OfferTimeTo'], format ='%I:%M %p').dt.strftime('%H:%M')



timeFrom = [i for i in df['OfferTimeFrom']]
        
timeTo = [j for j in df['OfferTimeTo']]


timeDict = {"COURSEID":courseID,"TIMEFROM":timeFrom, "TIMETO":timeTo}

        

######################### CONNECT TO SQLITE #################################

conn = sqlite3.connect('../data/classroom.db')

c = conn.cursor()

######################## INSERT TABLE TO DATABASE ############################
tables = """
    drop table if exists instructor;
    CREATE TABLE INSTRUCTOR(
       INSTRUCTORID INT PRIMARY KEY     NOT NULL,
       INSTRUCTORNAME           TEXT    
    );

    drop table if exists course;
    CREATE TABLE COURSE(
       COURSEID INT PRIMARY KEY     NOT NULL,
       COURSECODE TEXT,
       COURSENAME           TEXT    
    );

    drop table if exists room;
    CREATE TABLE ROOM(
       INSTRUCTORID INT      NOT NULL,
       COURSEID           INT,
       ROOMNAME TEXT,
       DAY TEXT
    );

    drop table if exists time;
    CREATE TABLE TIME(
       COURSEID           TEXT,
       TIMEFROM TIME (0) NOT NULL, 
       TIMETO TIME (0) NOT NULL
    );
    """

c.executescript(tables)

######################## INSERT DATA INTO TABLES ################################

#Insert Instructor Table
df = DataFrame(instructorDict, columns= ['INSTRUCTORID', 'INSTRUCTORNAME'])
df.to_sql('INSTRUCTOR', conn, if_exists='replace', index = False)

#Insert Course Table
df = DataFrame(courseDict, columns= ['COURSEID', 'COURSECODE','COURSENAME'])
df.to_sql('COURSE', conn, if_exists='replace', index = False)

#Insert Room Table
df = DataFrame(roomDict, columns= ['INSTRUCTORID','COURSEID','ROOMNAME','DAYS'])
df.to_sql('ROOM', conn, if_exists='replace', index = False)


#Insert Time Table
df = DataFrame(timeDict, columns= ['COURSEID','TIMEFROM','TIMETO'])
df.to_sql('TIME', conn, if_exists='replace', index = False)


conn.commit()
conn.close()



```
